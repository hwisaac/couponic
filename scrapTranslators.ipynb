```python
import json
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

def formNum(n):
    return "{:03d}".format(n)

# Chrome 옵션 설정
options = Options()
options.set_capability('goog:loggingPrefs', {'performance': 'ALL'})

# WebDriver 서비스 설정
service = Service()

# Chrome WebDriver 인스턴스 생성
driver = webdriver.Chrome(service=service, options=options)
driver.set_window_size(1500, 800)
driver.get("http://translators.co.kr/translatorIntro/detail?member_id=luvmeag")

def save_file(i):
  # 네트워크 로그 추출
  logs = driver.get_log("performance")
  response_body = None

  for entry in logs:
      log = json.loads(entry["message"])["message"]
      if log["method"] == "Network.responseReceived" and "response" in log["params"]:
          response_url = log["params"]["response"]["url"]
          if "http://translators.co.kr/translator/list/detailView" in response_url:
              request_id = log["params"]["requestId"]
              # 응답 본문 가져오기
              try:
                  response = driver.execute_cdp_cmd("Network.getResponseBody", {"requestId": request_id})
                  # 바로 JSON 객체로 변환
                  response_body = json.loads(response['body'])
              except Exception as e:
                  print(f"Error getting response body: {e}")
                  continue

  # 응답 본문을 JSON 객체로 저장
  if response_body:
      # 파일로 저장
      with open(f"./mnt/data/{formNum(i)}.json", "w", encoding="utf-8") as file:
          json.dump(response_body["result"], file, ensure_ascii=False, indent=4)
      print("파일이 성공적으로 저장되었습니다. /mnt/data/response_body.json")
  else:
      print("No response body for the specified URL.")



# 페이지가 완전히 로드될 때까지 충분한 시간 기다림
time.sleep(2)
initNum = 131
save_file(initNum)

time.sleep(2)

for i in range(initNum + 1,355):
  driver.find_element(By.CSS_SELECTOR, '#next').click()
  time.sleep(2)
  save_file(i)
driver.quit()
# for i in range(10):
#   time.sleep(2)
#   save_file(i)
```
